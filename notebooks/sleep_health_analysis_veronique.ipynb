```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yaml
```

```python
try:
    with open("../config.yaml", "r") as file:
        config = yaml.safe_load(file)
except:
    print("Configuration file not found!")
```

```python
sleep_df_clean = pd.read_csv(config['output_data']['file'], encoding='ISO-8859-1')
sleep_df_clean.head(5)
```

```python
"""
proportion of obese in sleep apnea
"""
proportion_table_by_bmi = pd.crosstab(sleep_df_clean['bmi_category'], sleep_df_clean['sleep_disorder'], normalize='index') * 100

print("Proportion of obese per sleep apnea ( %) :")
print(proportion_table_by_bmi.round(1))
```

# Conclusion H1a: Obesity increases likelihood of sleep apnea
The table indicates that obese people are more affected by both insomnia and sleep apnea than the rest of the population.
Furthermore, the data suggests that in this sample, no obese person is entirely free of a sleep disorder

```python
proportion_table_stress = pd.crosstab(sleep_df_clean['stress_level'], sleep_df_clean['sleep_disorder'], normalize='index') * 100

print("Proportion of stress per insomnia ( %) :")
print(proportion_table_stress.round(1))
```

# Conclusion H1b: Higher stress correlates with insomnia.

This table does not show correlation between higher stress and indomnia. Furthermore, individuals with no sleep disorder have significantly higher stress levels than those with insomnia.

```python
sleep_df_clean['has_disorder'] = (
    (sleep_df_clean['diastolic'] > 90) | 
    (sleep_df_clean['bmi_category'] == 'obese') | 
    (sleep_df_clean['systolic'] > 140) |
    (sleep_df_clean['heart_rate'] > 100) |
    (sleep_df_clean['heart_rate'] < 60) |
    (sleep_df_clean['sleep_disorder'] != 'No Disorder')
    ).astype(int)
```

```python

proportion_table_duration_sleep = pd.crosstab(sleep_df_clean['sleep_duration'] < 6, sleep_df_clean['has_disorder'] == 1, normalize='index') * 100

print("Proportion of duration sleep < 6h increase disorder ( %) :")
print(proportion_table_duration_sleep.round(1))

```
